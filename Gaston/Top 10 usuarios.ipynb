```python
# imports de librerias
import pandas as pd
```

```python
# Leo el archivo de auctions.
auctions = pd.read_csv('auctions.csv.gzip', compression='gzip')
auctions.head(20)

# Leo el archivo de installs.
installs = pd.read_csv('installs.csv.gzip', compression='gzip')
installs.head(20)

# Leo el archivo de eventos.
events = pd.read_csv('events.csv.gzip', compression='gzip')
events.head(20)

# Leo el archivo de clicks.
clicks = pd.read_csv('clicks.csv.gzip', compression='gzip')
clicks.head(20)
```

```python
# Top 10 de cantidad de auctions por usuario.       
auctionsPerUser = auctions[['device_id']]
auctionsPerUser.rename(columns={'device_id': 'ref_hash'}, inplace=True)
auctionsPerUser['auctionsCount'] = 1
auctionsPerUser = auctionsPerUser.groupby('ref_hash').count()
auctionsPerUser = auctionsPerUser.sort_values(by='auctionsCount', ascending=False)
auctionsPerUser.reset_index(inplace=True)
auctionsPerUserTop10 = auctionsPerUser.head(10)
auctionsPerUserTop10
```

```python
# Cantidad de installs por usuario.
installsPerUser = installs[['ref_hash']]
installsPerUser['installsCount'] = 1
installsPerUser = installsPerUser.groupby('ref_hash').count()
installsPerUser = installsPerUser.sort_values(by='installsCount', ascending=False)
installsPerUser.reset_index(inplace=True)
installsPerUserTop10 = installsPerUser.head(10)
installsPerUserTop10
```

```python
# Cantidad de eventos por usuario.
eventsPerUser = events[['ref_hash']]
eventsPerUser['eventsCount'] = 1
eventsPerUser = eventsPerUser.groupby('ref_hash').count()
eventsPerUser = eventsPerUser.sort_values(by='eventsCount', ascending=False)
eventsPerUser.reset_index(inplace=True)
eventsPerUserTop10 = eventsPerUser.head(10)
eventsPerUserTop10
```

```python
# Cantidad de clicks por usuario.
clicksPerUser = clicks[['ref_hash']]
clicksPerUser['clicksCount'] = 1
clicksPerUser = clicksPerUser.groupby('ref_hash').count()
clicksPerUser = clicksPerUser.sort_values(by='clicksCount', ascending=False)
clicksPerUser.reset_index(inplace=True)
clicksPerUserTop10 = clicksPerUser.head(10)
clicksPerUserTop10
```

# Top 10 de usuarios con mas auctions vs installs vs events vs clicks.

```python
# Mergeo todos los dataframes que me quedaron con el de auctionsPerUserTop10.
mergedDataTopAuctions = pd.merge(auctionsPerUserTop10, installsPerUser, on='ref_hash', how='left')
mergedDataTopAuctions = pd.merge(mergedDataTopAuctions, eventsPerUser, on='ref_hash', how='left')
mergedDataTopAuctions = pd.merge(mergedDataTopAuctions, clicksPerUser, on='ref_hash', how='left')
mergedDataTopAuctions.fillna(0)
```

# Top 10 de usuarios con mas installs vs auctions vs events vs clicks.

```python
# Mergeo todos los dataframes que me quedaron con el de installsPerUserTop10.
mergedDataTopInstalls = pd.merge(installsPerUserTop10, auctionsPerUser, on='ref_hash', how='left')
mergedDataTopInstalls = pd.merge(mergedDataTopInstalls, eventsPerUser, on='ref_hash', how='left')
mergedDataTopInstalls = pd.merge(mergedDataTopInstalls, clicksPerUser, on='ref_hash', how='left')
mergedDataTopInstalls.fillna(0)
```

# Top 10 de usuarios con mas eventos vs auctions vs installs vs clicks.

```python
# Mergeo todos los dataframes que me quedaron con el de eventsPerUserTop10.
mergedDataTopEvents = pd.merge(eventsPerUserTop10, installsPerUser, on='ref_hash', how='left')
mergedDataTopEvents = pd.merge(mergedDataTopEvents, auctionsPerUser, on='ref_hash', how='left')
mergedDataTopEvents = pd.merge(mergedDataTopEvents, clicksPerUser, on='ref_hash', how='left')
mergedDataTopEvents.fillna(0)
```

# Top 10 de usuarios con mas eventos vs auctions vs installs vs clicks.

```python
# Mergeo todos los dataframes que me quedaron con el de clicksPerUserTop10.
mergedDataTopClicks = pd.merge(clicksPerUserTop10, installsPerUser, on='ref_hash', how='left')
mergedDataTopClicks = pd.merge(mergedDataTopClicks, auctionsPerUser, on='ref_hash', how='left')
mergedDataTopClicks = pd.merge(mergedDataTopClicks, eventsPerUser, on='ref_hash', how='left')
mergedDataTopClicks.fillna(0)
```
