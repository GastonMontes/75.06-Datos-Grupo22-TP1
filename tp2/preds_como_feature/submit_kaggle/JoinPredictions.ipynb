```python
import pandas as pd
import numpy as np
```

```python
pred_installs = pd.read_csv('pred_installs.csv')
pred_auctions = pd.read_csv('pred_auctions.csv')

pred_installs.columns = ['ref_hash', 'pred_inst']
pred_auctions.columns = ['ref_hash', 'pred_auct']

targets = pd.read_csv('../target_competencia_ids.csv')
```

```python
targets = targets.merge(pred_installs, on='ref_hash', how='left').drop('obj', axis=1)
targets
```

```python
targets = targets.merge(pred_auctions, on='ref_hash', how='left')
```

```python
targets.fillna(0, inplace=True)
```

```python
targets['preds'] = targets['pred_inst'] + targets['pred_auct']
targets
```

```python
del targets['pred_inst']
del targets['pred_auct']
targets
```

```python
targets.columns = ['ref_hash', 'obj']
```

```python
targets.to_csv('submit_kaggle.csv', index=False)
```
