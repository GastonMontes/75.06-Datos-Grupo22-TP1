```python
import pandas as pd

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

# Read the data.

```python
path_18_20 = 'datos/test_and_train/auctions_18_20.csv'
auctions_18_20 = pd.read_csv(path_18_20)

path_19_21 = 'datos/test_and_train/auctions_19_21.csv'
auctions_19_21 = pd.read_csv(path_19_21)

path_20_22 = 'datos/test_and_train/auctions_20_22.csv'
auctions_20_22 = pd.read_csv(path_20_22)

path_21_23 = 'datos/test_and_train/auctions_21_23.csv'
auctions_21_23 = pd.read_csv(path_21_23)

path_22_24 = 'datos/test_and_train/auctions_22_24.csv'
auctions_22_24 = pd.read_csv(path_22_24)

path_23_25 = 'datos/test_and_train/auctions_23_25.csv'
auctions_23_25 = pd.read_csv(path_23_25)

path_24_26 = 'datos/test_and_train/auctions_24_26.csv'
auctions_24_26 = pd.read_csv(path_24_26)
```

# Clear the data.

## Set values to NaN.

```python
auctions_18_20['ref_hash'].fillna(0, inplace= True)
auctions_18_20['cant_auctions'].fillna(0, inplace= True)
auctions_18_20['cant_sources'].fillna(0, inplace= True)
auctions_18_20['cant_types'].fillna(0, inplace= True)
auctions_18_20['time_to_auction'].fillna(259200, inplace= True)

auctions_19_21['ref_hash'].fillna(0, inplace= True)
auctions_19_21['cant_auctions'].fillna(0, inplace= True)
auctions_19_21['cant_sources'].fillna(0, inplace= True)
auctions_19_21['cant_types'].fillna(0, inplace= True)
auctions_19_21['time_to_auction'].fillna(259200, inplace= True)

auctions_20_22['ref_hash'].fillna(0, inplace= True)
auctions_20_22['cant_auctions'].fillna(0, inplace= True)
auctions_20_22['cant_sources'].fillna(0, inplace= True)
auctions_20_22['cant_types'].fillna(0, inplace= True)
auctions_20_22['time_to_auction'].fillna(259200, inplace= True)

auctions_21_23['ref_hash'].fillna(0, inplace= True)
auctions_21_23['cant_auctions'].fillna(0, inplace= True)
auctions_21_23['cant_sources'].fillna(0, inplace= True)
auctions_21_23['cant_types'].fillna(0, inplace= True)
auctions_21_23['time_to_auction'].fillna(259200, inplace= True)

auctions_22_24['ref_hash'].fillna(0, inplace= True)
auctions_22_24['cant_auctions'].fillna(0, inplace= True)
auctions_22_24['cant_sources'].fillna(0, inplace= True)
auctions_22_24['cant_types'].fillna(0, inplace= True)
auctions_22_24['time_to_auction'].fillna(259200, inplace= True)

auctions_23_25['ref_hash'].fillna(0, inplace= True)
auctions_23_25['cant_auctions'].fillna(0, inplace= True)
auctions_23_25['cant_sources'].fillna(0, inplace= True)
auctions_23_25['cant_types'].fillna(0, inplace= True)
auctions_23_25['time_to_auction'].fillna(259200, inplace= True)

auctions_24_26['ref_hash'].fillna(0, inplace= True)
auctions_24_26['cant_auctions'].fillna(0, inplace= True)
auctions_24_26['cant_sources'].fillna(0, inplace= True)
auctions_24_26['cant_types'].fillna(0, inplace= True)
auctions_24_26['time_to_auction'].fillna(259200, inplace= True)
```

# Define train dataset, test dataset and stats.

## Train dataset.

```python
train_dataset_18_20 = auctions_18_20.copy()
train_dataset_19_21 = auctions_19_21.copy()
train_dataset_20_22 = auctions_20_22.copy()
train_dataset_21_23 = auctions_21_23.copy()
train_dataset_22_24 = auctions_22_24.copy()
train_dataset_23_25 = auctions_23_25.copy()
```

## Test dataset.

```python
test_dataset_18_20 = auctions_19_21.copy()
test_dataset_19_21 = auctions_20_22.copy()
test_dataset_20_22 = auctions_21_23.copy()
test_dataset_21_23 = auctions_22_24.copy()
test_dataset_22_24 = auctions_23_25.copy()
test_dataset_23_25 = auctions_24_26.copy()
```

## Stats.

```python
train_stats_18_20 = train_dataset_18_20.describe()
train_stats_18_20.pop('time_to_auction')
train_stats_18_20 = train_stats_18_20.transpose()

train_stats_19_21 = train_dataset_19_21.describe()
train_stats_19_21.pop('time_to_auction')
train_stats_19_21 = train_stats_19_21.transpose()

train_stats_20_22 = train_dataset_20_22.describe()
train_stats_20_22.pop('time_to_auction')
train_stats_20_22 = train_stats_20_22.transpose()

train_stats_21_23 = train_dataset_21_23.describe()
train_stats_21_23.pop('time_to_auction')
train_stats_21_23 = train_stats_21_23.transpose()

train_stats_22_24 = train_dataset_22_24.describe()
train_stats_22_24.pop('time_to_auction')
train_stats_22_24 = train_stats_22_24.transpose()

train_stats_23_25 = train_dataset_23_25.describe()
train_stats_23_25.pop('time_to_auction')
train_stats_23_25 = train_stats_23_25.transpose()
```

## Define labels.

```python
train_labels_18_20 = train_dataset_18_20.pop('time_to_auction')
test_labels_18_20 = test_dataset_18_20.pop('time_to_auction')

train_labels_19_21 = train_dataset_19_21.pop('time_to_auction')
test_labels_19_21 = test_dataset_19_21.pop('time_to_auction')

train_labels_20_22 = train_dataset_20_22.pop('time_to_auction')
test_labels_20_22 = test_dataset_20_22.pop('time_to_auction')

train_labels_21_23 = train_dataset_21_23.pop('time_to_auction')
test_labels_21_23 = test_dataset_21_23.pop('time_to_auction')

train_labels_22_24 = train_dataset_22_24.pop('time_to_auction')
test_labels_22_24 = test_dataset_22_24.pop('time_to_auction')

train_labels_23_25 = train_dataset_23_25.pop('time_to_auction')
test_labels_23_25 = test_dataset_23_25.pop('time_to_auction')
```

## Normalize train and test datasets.

```python
def norm(stats, x):
  return (x - stats['mean']) / stats['std']

train_dataset_18_20 = norm(train_stats_18_20, train_dataset_18_20)
test_dataset_18_20 = norm(train_stats_18_20, test_dataset_18_20)

train_dataset_19_21 = norm(train_stats_19_21, train_dataset_19_21)
test_dataset_19_21 = norm(train_stats_19_21, test_dataset_19_21)

train_dataset_20_22 = norm(train_stats_20_22, train_dataset_20_22)
test_dataset_20_22 = norm(train_stats_20_22, test_dataset_20_22)

train_dataset_21_23 = norm(train_stats_21_23, train_dataset_21_23)
test_dataset_21_23 = norm(train_stats_21_23, test_dataset_21_23)

train_dataset_22_24 = norm(train_stats_22_24, train_dataset_22_24)
test_dataset_22_24 = norm(train_stats_22_24, test_dataset_22_24)

train_dataset_23_25 = norm(train_stats_23_25, train_dataset_23_25)
test_dataset_23_25 = norm(train_stats_23_25, test_dataset_23_25)
```

# Build the model.

```python
def build_model(): 
    model = keras.Sequential([
        layers.Dense(64, activation=tf.nn.relu, input_shape=[len(train_dataset_18_20.keys())]),
        layers.Dense(64, activation=tf.nn.relu),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mean_squared_error',
                  optimizer=optimizer,
                  metrics=['mean_absolute_error', 'mean_squared_error'])
    return model

model_18_20 = build_model()
model_19_21 = build_model()
model_20_22 = build_model()
model_21_23 = build_model()
model_22_24 = build_model()
model_23_25 = build_model()
```

## Train the model.

```python
EPOCHS = 10000

class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')
    
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=100)

history_18_20 = model_18_20.fit(train_dataset_18_20, train_labels_18_20, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

history_19_21 = model_19_21.fit(train_dataset_19_21, train_labels_19_21, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

history_20_22 = model_20_22.fit(train_dataset_20_22, train_labels_20_22, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

history_21_23 = model_21_23.fit(train_dataset_21_23, train_labels_21_23, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

history_22_24 = model_22_24.fit(train_dataset_22_24, train_labels_22_24, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

history_23_25 = model_23_25.fit(train_dataset_23_25, train_labels_23_25, epochs=EPOCHS, 
                                validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])
```

```python
loss_18_20, mae_18_20, mse_18_20 = model_18_20.evaluate(test_dataset_18_20, test_labels_18_20, verbose=0)
print("Testing set Mean Abs Error _18_20: {:5.5f}".format(mae_18_20))
print("Testing set Mean square Error _18_20: {:5.5f}".format(mse_18_20))

loss_19_21, mae_19_21, mse_19_21 = model_19_21.evaluate(test_dataset_19_21, test_labels_19_21, verbose=0)
print("Testing set Mean Abs Error _19_21: {:5.2f}".format(mae_19_21))
print("Testing set Mean square Error _19_21: {:5.2f}".format(mse_19_21))

loss_20_22, mae_20_22, mse_20_22 = model_20_22.evaluate(test_dataset_20_22, test_labels_20_22, verbose=0)
print("Testing set Mean Abs Error _20_22: {:5.2f}".format(mae_20_22))
print("Testing set Mean square Error _20_22: {:5.2f}".format(mse_20_22))

loss_21_23, mae_21_23, mse_21_23 = model_21_23.evaluate(test_dataset_21_23, test_labels_21_23, verbose=0)
print("Testing set Mean Abs Error _21_23: {:5.2f}".format(mae_21_23))
print("Testing set Mean square Error _21_23: {:5.2f}".format(mse_21_23))

loss_22_24, mae_22_24, mse_22_24 = model_22_24.evaluate(test_dataset_22_24, test_labels_22_24, verbose=0)
print("Testing set Mean Abs Error _22_24: {:5.2f}".format(mae_22_24))
print("Testing set Mean square Error _22_24: {:5.2f}".format(mse_22_24))

loss_23_25, mae_23_25, mse_23_25 = model_23_25.evaluate(test_dataset_23_25, test_labels_23_25, verbose=0)
print("Testing set Mean Abs Error _23_25: {:5.2f}".format(mae_23_25))
print("Testing set Mean square Error _23_25: {:5.2f}".format(mse_23_25))
```

## Make predictions

```python
test_predictions_18_20 = model_18_20.predict(test_dataset_18_20)
test_predictions_19_21 = model_19_21.predict(test_dataset_19_21)
test_predictions_20_22 = model_20_22.predict(test_dataset_20_22)
test_predictions_21_23 = model_21_23.predict(test_dataset_21_23)
test_predictions_22_24 = model_22_24.predict(test_dataset_22_24)
test_predictions_23_25 = model_23_25.predict(test_dataset_23_25)
```

```python
pred_DF = pd.DataFrame(auctions_18_20['ref_hash'])

pred_DF["pred_18_20"] = test_predictions_18_20
pred_DF["value_18_20"] = test_labels_18_20
pred_DF['pred_18_20'].fillna(259200, inplace= True)

pred_DF["pred_19_21"] = test_predictions_19_21
pred_DF["value_19_21"] = test_labels_19_21
pred_DF['pred_19_21'].fillna(259200, inplace= True)

pred_DF["pred_20_22"] = test_predictions_20_22
pred_DF["value_20_22"] = test_labels_20_22
pred_DF['pred_20_22'].fillna(259200, inplace= True)

pred_DF["pred_21_23"] = test_predictions_21_23
pred_DF["value_21_23"] = test_labels_21_23
pred_DF['pred_21_23'].fillna(259200, inplace= True)

pred_DF["pred_22_24"] = test_predictions_22_24
pred_DF["value_22_24"] = test_labels_22_24
pred_DF['pred_22_24'].fillna(259200, inplace= True)

pred_DF["pred_23_25"] = test_predictions_23_25
pred_DF["value_23_25"] = test_labels_23_25
pred_DF['pred_23_25'].fillna(259200, inplace= True)

pred_DF.head(50)
```

```python
predictions = pd.DataFrame(pred_DF['ref_hash'])
predictions['obj'] = (pred_DF['pred_18_20'] + pred_DF['pred_19_21'] + pred_DF['pred_20_22'] + pred_DF['pred_21_23'] + pred_DF['pred_22_24'] + pred_DF['pred_23_25']) / 6

predictions['ref_hash'] = predictions['ref_hash'].astype(str) + '_st'
predictions['ref_hash'] = predictions['ref_hash'].astype('object')
```

## Save predictions.

```python
predictions.to_csv('datos/predictions/auctions.csv', index=False)
```
