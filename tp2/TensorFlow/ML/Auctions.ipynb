```python
import pandas as pd
import numpy as np
import datetime
```

# Leo el archivo de auctions.

```python
path_auctions = 'datos/auctions.csv'
auctions = pd.read_csv(path_auctions)
auctions.head()
```

# Se da formato a las fechas.

```python
# Damos formato a la fecha.
auctions['timestamp_format'] = pd.to_datetime(auctions['date'])

# Agrego columna fecha.
auctions['date_timestamp'] = auctions['timestamp_format'].dt.date

# Agrego columna numero dia.
auctions['day'] = auctions['timestamp_format'].dt.day

auctions['ref_hash'] = auctions['device_id']
del auctions['device_id']

auctions.head()
```

# Se crean las ventanas.

```python
auctions18_20 = auctions[(auctions['day'] >= 18) & (auctions['day'] < 21)]
auctions19_21 = auctions[(auctions['day'] >= 19) & (auctions['day'] < 22)]
auctions20_22 = auctions[(auctions['day'] >= 20) & (auctions['day'] < 23)]
auctions21_23 = auctions[(auctions['day'] >= 21) & (auctions['day'] < 24)]
auctions22_24 = auctions[(auctions['day'] >= 22) & (auctions['day'] < 25)]
auctions23_25 = auctions[(auctions['day'] >= 23) & (auctions['day'] < 26)]
auctions24_26 = auctions[(auctions['day'] >= 24) & (auctions['day'] < 27)]
auctions23_25.head()
```

# Se agrega el time to auction.

```python
auctions18_20['base_time'] = datetime.datetime(2019, 4, 18)
auctions18_20['time_to_auction'] = (auctions18_20['timestamp_format'] - auctions18_20['base_time']) / np.timedelta64(1, 's')

auctions19_21['base_time'] = datetime.datetime(2019, 4, 19)
auctions19_21['time_to_auction'] = (auctions19_21['timestamp_format'] - auctions19_21['base_time']) / np.timedelta64(1, 's')

auctions20_22['base_time'] = datetime.datetime(2019, 4, 20)
auctions20_22['time_to_auction'] = (auctions20_22['timestamp_format'] - auctions20_22['base_time']) / np.timedelta64(1, 's')

auctions21_23['base_time'] = datetime.datetime(2019, 4, 21)
auctions21_23['time_to_auction'] = (auctions21_23['timestamp_format'] - auctions21_23['base_time']) / np.timedelta64(1, 's')

auctions22_24['base_time'] = datetime.datetime(2019, 4, 22)
auctions22_24['time_to_auction'] = (auctions22_24['timestamp_format'] - auctions22_24['base_time']) / np.timedelta64(1, 's')

auctions23_25['base_time'] = datetime.datetime(2019, 4, 23)
auctions23_25['time_to_auction'] = (auctions23_25['timestamp_format'] - auctions23_25['base_time']) / np.timedelta64(1, 's')

auctions24_26['base_time'] = datetime.datetime(2019, 4, 24)
auctions24_26['time_to_auction'] = (auctions24_26['timestamp_format'] - auctions24_26['base_time']) / np.timedelta64(1, 's')

auctions21_23.head()
```

# Se agrega la cantidad de subastas.

```python
subastas_18_20 = auctions18_20.groupby('ref_hash')['date'].count().reset_index()
subastas_18_20.columns = ['ref_hash', 'cant_auctions']

subastas_19_21 = auctions19_21.groupby('ref_hash')['date'].count().reset_index()
subastas_19_21.columns = ['ref_hash', 'cant_auctions']

subastas_20_22 = auctions20_22.groupby('ref_hash')['date'].count().reset_index()
subastas_20_22.columns = ['ref_hash', 'cant_auctions']

subastas_21_23 = auctions21_23.groupby('ref_hash')['date'].count().reset_index()
subastas_21_23.columns = ['ref_hash', 'cant_auctions']

subastas_22_24 = auctions22_24.groupby('ref_hash')['date'].count().reset_index()
subastas_22_24.columns = ['ref_hash', 'cant_auctions']

subastas_23_25 = auctions23_25.groupby('ref_hash')['date'].count().reset_index()
subastas_23_25.columns = ['ref_hash', 'cant_auctions']

subastas_24_26 = auctions24_26.groupby('ref_hash')['date'].count().reset_index()
subastas_24_26.columns = ['ref_hash', 'cant_auctions']

subastas_22_24.head()
```

# Se agrega la cantidad de sources.

```python
sources_18_20 = auctions18_20.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_18_20.columns = ['ref_hash', 'cant_sources']

sources_19_21 = auctions19_21.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_19_21.columns = ['ref_hash', 'cant_sources']

sources_20_22 = auctions20_22.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_20_22.columns = ['ref_hash', 'cant_sources']

sources_21_23 = auctions21_23.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_21_23.columns = ['ref_hash', 'cant_sources']

sources_22_24 = auctions22_24.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_22_24.columns = ['ref_hash', 'cant_sources']

sources_23_25 = auctions23_25.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_23_25.columns = ['ref_hash', 'cant_sources']

sources_24_26 = auctions24_26.groupby('ref_hash')['source_id'].nunique().reset_index()
sources_24_26.columns = ['ref_hash', 'cant_sources']

sources_18_20.head()
```

```python
subastas_18_20 = subastas_18_20.merge(sources_18_20, on='ref_hash', how='left')
subastas_19_21 = subastas_19_21.merge(sources_18_20, on='ref_hash', how='left')
subastas_20_22 = subastas_20_22.merge(sources_18_20, on='ref_hash', how='left')
subastas_21_23 = subastas_21_23.merge(sources_18_20, on='ref_hash', how='left')
subastas_22_24 = subastas_22_24.merge(sources_18_20, on='ref_hash', how='left')
subastas_23_25 = subastas_23_25.merge(sources_18_20, on='ref_hash', how='left')
subastas_24_26 = subastas_24_26.merge(sources_18_20, on='ref_hash', how='left')

subastas_19_21.head()
```

```python
subastas_18_20['cant_sources'].fillna(0, inplace= True)
subastas_18_20['cant_auctions'].fillna(0, inplace= True)

subastas_19_21['cant_sources'].fillna(0, inplace= True)
subastas_19_21['cant_auctions'].fillna(0, inplace= True)


subastas_20_22['cant_sources'].fillna(0, inplace= True)
subastas_20_22['cant_auctions'].fillna(0, inplace= True)


subastas_21_23['cant_sources'].fillna(0, inplace= True)
subastas_21_23['cant_auctions'].fillna(0, inplace= True)


subastas_22_24['cant_sources'].fillna(0, inplace= True)
subastas_22_24['cant_auctions'].fillna(0, inplace= True)


subastas_23_25['cant_sources'].fillna(0, inplace= True)
subastas_23_25['cant_auctions'].fillna(0, inplace= True)


subastas_24_26['cant_sources'].fillna(0, inplace= True)
subastas_24_26['cant_auctions'].fillna(0, inplace= True)

subastas_20_22.head()
```

# Se agrega la cantidad de reference types.

```python
types_18_20 = auctions18_20.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_18_20.columns = ['ref_hash', 'cant_types']

types_19_21 = auctions19_21.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_19_21.columns = ['ref_hash', 'cant_types']

types_20_22 = auctions20_22.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_20_22.columns = ['ref_hash', 'cant_types']

types_21_23 = auctions21_23.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_21_23.columns = ['ref_hash', 'cant_types']

types_22_24 = auctions22_24.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_22_24.columns = ['ref_hash', 'cant_types']

types_23_25 = auctions23_25.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_23_25.columns = ['ref_hash', 'cant_types']

types_24_26 = auctions24_26.groupby('ref_hash')['ref_type_id'].nunique().reset_index()
types_24_26.columns = ['ref_hash', 'cant_types']

types_21_23.head()
```

```python
subastas_18_20 = subastas_18_20.merge(types_18_20, on='ref_hash', how='left')
subastas_19_21 = subastas_19_21.merge(types_19_21, on='ref_hash', how='left')
subastas_20_22 = subastas_20_22.merge(types_20_22, on='ref_hash', how='left')
subastas_21_23 = subastas_21_23.merge(types_21_23, on='ref_hash', how='left')
subastas_22_24 = subastas_22_24.merge(types_22_24, on='ref_hash', how='left')
subastas_23_25 = subastas_23_25.merge(types_23_25, on='ref_hash', how='left')
subastas_24_26 = subastas_24_26.merge(types_24_26, on='ref_hash', how='left')

subastas_22_24.head()
```

```python
subastas_18_20['cant_types'].fillna(0, inplace= True)
subastas_19_21['cant_types'].fillna(0, inplace= True)
subastas_20_22['cant_types'].fillna(0, inplace= True)
subastas_21_23['cant_types'].fillna(0, inplace= True)
subastas_22_24['cant_types'].fillna(0, inplace= True)
subastas_23_25['cant_types'].fillna(0, inplace= True)
subastas_24_26['cant_types'].fillna(0, inplace= True)

subastas_23_25.head()
```

# Se agrega el time to auction.

```python
time_to_auction_18_20 = auctions18_20.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_18_20.columns = ['ref_hash', 'time_to_auction']

time_to_auction_19_21 = auctions19_21.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_19_21.columns = ['ref_hash', 'time_to_auction']

time_to_auction_20_22 = auctions20_22.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_20_22.columns = ['ref_hash', 'time_to_auction']

time_to_auction_21_23 = auctions21_23.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_21_23.columns = ['ref_hash', 'time_to_auction']

time_to_auction_22_24 = auctions22_24.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_22_24.columns = ['ref_hash', 'time_to_auction']

time_to_auction_23_25 = auctions23_25.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_23_25.columns = ['ref_hash', 'time_to_auction']

time_to_auction_24_26 = auctions24_26.groupby('ref_hash').agg({'time_to_auction': 'mean'}).reset_index()
time_to_auction_24_26.columns = ['ref_hash', 'time_to_auction']

time_to_auction_24_26.head()
```

```python
subastas_18_20 = subastas_18_20.merge(time_to_auction_18_20, on='ref_hash', how='left')
subastas_19_21 = subastas_19_21.merge(time_to_auction_19_21, on='ref_hash', how='left')
subastas_20_22 = subastas_20_22.merge(time_to_auction_20_22, on='ref_hash', how='left')
subastas_21_23 = subastas_21_23.merge(time_to_auction_21_23, on='ref_hash', how='left')
subastas_22_24 = subastas_22_24.merge(time_to_auction_22_24, on='ref_hash', how='left')
subastas_23_25 = subastas_23_25.merge(time_to_auction_23_25, on='ref_hash', how='left')
subastas_24_26 = subastas_24_26.merge(time_to_auction_24_26, on='ref_hash', how='left')

subastas_18_20.head()
```

```python
# 3 dias son 259200 segundos
subastas_18_20['time_to_auction'].fillna(259200, inplace= True)
subastas_19_21['time_to_auction'].fillna(259200, inplace= True)
subastas_20_22['time_to_auction'].fillna(259200, inplace= True)
subastas_21_23['time_to_auction'].fillna(259200, inplace= True)
subastas_22_24['time_to_auction'].fillna(259200, inplace= True)
subastas_23_25['time_to_auction'].fillna(259200, inplace= True)
subastas_24_26['time_to_auction'].fillna(259200, inplace= True)

subastas_19_21.head()
```

# Join de targets con features.

```python
# Leo el archivo target.
targets = pd.read_csv('datos/target_competencia_ids.csv')

targets['ref_hash'].replace(regex = True, inplace = True, to_replace='_sc',value='')
targets['ref_hash'].replace(regex = True, inplace = True, to_replace='_st',value='')

targets['ref_hash'] = targets['ref_hash'].astype(str).astype(int)

targets = targets.drop_duplicates(subset='ref_hash', keep='first')


targets.head()
```

```python
merged_targets_18_20 = targets.merge(subastas_18_20, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_19_21 = targets.merge(subastas_19_21, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_20_22 = targets.merge(subastas_20_22, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_21_23 = targets.merge(subastas_21_23, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_22_24 = targets.merge(subastas_22_24, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_23_25 = targets.merge(subastas_23_25, on='ref_hash', how='left').drop('obj', axis=1)
merged_targets_24_26 = targets.merge(subastas_24_26, on='ref_hash', how='left').drop('obj', axis=1)

merged_targets_21_23.head()
```

```python
# 3 dias son 259200 segundos
merged_targets_18_20['time_to_auction'].fillna(259200, inplace= True)
merged_targets_19_21['time_to_auction'].fillna(259200, inplace= True)
merged_targets_20_22['time_to_auction'].fillna(259200, inplace= True)
merged_targets_21_23['time_to_auction'].fillna(259200, inplace= True)
merged_targets_22_24['time_to_auction'].fillna(259200, inplace= True)
merged_targets_23_25['time_to_auction'].fillna(259200, inplace= True)
merged_targets_24_26['time_to_auction'].fillna(259200, inplace= True)

merged_targets_22_24.head()
```

# Se guardan los datos.

```python
merged_targets_18_20.to_csv('datos/test_and_train/auctions_18_20.csv', index=False)
merged_targets_19_21.to_csv('datos/test_and_train/auctions_19_21.csv', index=False)
merged_targets_20_22.to_csv('datos/test_and_train/auctions_20_22.csv', index=False)
merged_targets_21_23.to_csv('datos/test_and_train/auctions_21_23.csv', index=False)
merged_targets_22_24.to_csv('datos/test_and_train/auctions_22_24.csv', index=False)
merged_targets_23_25.to_csv('datos/test_and_train/auctions_23_25.csv', index=False)
merged_targets_24_26.to_csv('datos/test_and_train/auctions_24_26.csv', index=False)
```
