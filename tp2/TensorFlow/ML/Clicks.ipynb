```python
import pandas as pd
import numpy as np
import datetime
```

# Leo el archivo de Clicks.

```python
path = 'datos/clicks.csv'
clicks = pd.read_csv(path)
clicks.head()
```

# Se da formato a las fechas.

```python
# Damos formato a la fecha.
clicks['timestamp_format'] = pd.to_datetime(clicks['created'])

# Agrego columna fecha.
clicks['date_timestamp'] = clicks['timestamp_format'].dt.date

# Agrego columna numero dia.
clicks['day'] = clicks['timestamp_format'].dt.day

clicks.head()
```

# Se crean las ventanas.

```python
clicks_18_20 = clicks[(clicks['day'] >= 18) & (clicks['day'] < 21)]
clicks_19_21 = clicks[(clicks['day'] >= 19) & (clicks['day'] < 22)]
clicks_20_22 = clicks[(clicks['day'] >= 20) & (clicks['day'] < 23)]
clicks_21_23 = clicks[(clicks['day'] >= 21) & (clicks['day'] < 24)]
clicks_22_24 = clicks[(clicks['day'] >= 22) & (clicks['day'] < 25)]
clicks_23_25 = clicks[(clicks['day'] >= 23) & (clicks['day'] < 26)]
clicks_24_26 = clicks[(clicks['day'] >= 24) & (clicks['day'] < 27)]

clicks18_20.head()
```

# Se agregan la cantidad de clicks.

```python
clicksDT_18_20 = clicks_18_20.groupby('ref_hash')['created'].count().reset_index()
clicksDT_18_20.columns = ['ref_hash', 'cant_clicks']

clicksDT_19_21 = clicks_19_21.groupby('ref_hash')['created'].count().reset_index()
clicksDT_19_21.columns = ['ref_hash', 'cant_clicks']

clicksDT_20_22 = clicks_20_22.groupby('ref_hash')['created'].count().reset_index()
clicksDT_20_22.columns = ['ref_hash', 'cant_clicks']

clicksDT_21_23 = clicks_21_23.groupby('ref_hash')['created'].count().reset_index()
clicksDT_21_23.columns = ['ref_hash', 'cant_clicks']

clicksDT_22_24 = clicks_22_24.groupby('ref_hash')['created'].count().reset_index()
clicksDT_22_24.columns = ['ref_hash', 'cant_clicks']

clicksDT_23_25 = clicks_23_25.groupby('ref_hash')['created'].count().reset_index()
clicksDT_23_25.columns = ['ref_hash', 'cant_clicks']

clicksDT_24_26 = clicks_24_26.groupby('ref_hash')['created'].count().reset_index()
clicksDT_24_26.columns = ['ref_hash', 'cant_clicks']

clicksDT_19_21.head()
```

```python
clicksDT_18_20['cant_clicks'].fillna(0, inplace= True)
clicksDT_19_21['cant_clicks'].fillna(0, inplace= True)
clicksDT_20_22['cant_clicks'].fillna(0, inplace= True)
clicksDT_21_23['cant_clicks'].fillna(0, inplace= True)
clicksDT_22_24['cant_clicks'].fillna(0, inplace= True)
clicksDT_23_25['cant_clicks'].fillna(0, inplace= True)
clicksDT_24_26['cant_clicks'].fillna(0, inplace= True)

clicksDT_20_22.head()
```

# Se agrega la cantidad de clicks con WI-FI.

```python
clicks_wifi_18_20 = clicks_18_20.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_18_20.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_19_21 = clicks_19_21.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_19_21.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_20_22 = clicks_20_22.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_20_22.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_21_23 = clicks_21_23.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_21_23.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_22_24 = clicks_22_24.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_22_24.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_23_25 = clicks_23_25.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_23_25.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_24_26 = clicks_24_26.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_wifi_24_26.columns = ['ref_hash', 'cant_wifi_clicks']

clicks_wifi_21_23.head()
```

```python
clicks_wifi_18_20['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_19_21['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_20_22['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_21_23['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_22_24['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_23_25['cant_wifi_clicks'].fillna(0, inplace= True)
clicks_wifi_24_26['cant_wifi_clicks'].fillna(0, inplace= True)

clicks_wifi_22_24.head()
```

```python
clicksDT_18_20 = clicksDT_18_20.merge(clicks_wifi_18_20, on='ref_hash', how='left')
clicksDT_19_21 = clicksDT_19_21.merge(clicks_wifi_19_21, on='ref_hash', how='left')
clicksDT_20_22 = clicksDT_20_22.merge(clicks_wifi_20_22, on='ref_hash', how='left')
clicksDT_21_23 = clicksDT_21_23.merge(clicks_wifi_21_23, on='ref_hash', how='left')
clicksDT_22_24 = clicksDT_22_24.merge(clicks_wifi_22_24, on='ref_hash', how='left')
clicksDT_23_25 = clicksDT_23_25.merge(clicks_wifi_23_25, on='ref_hash', how='left')
clicksDT_24_26 = clicksDT_24_26.merge(clicks_wifi_24_26, on='ref_hash', how='left')

clicksDT_23_25.head()
```

# Se agrega el promedio de tiempo de clicks.

```python
mean_click_18_20 = clicks_18_20.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_18_20.columns = ['ref_hash', 'mean_time_to_click']

mean_click_19_21 = clicks_19_21.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_19_21.columns = ['ref_hash', 'mean_time_to_click']

mean_click_20_22 = clicks_20_22.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_20_22.columns = ['ref_hash', 'mean_time_to_click']

mean_click_21_23 = clicks_21_23.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_21_23.columns = ['ref_hash', 'mean_time_to_click']

mean_click_22_24 = clicks_22_24.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_22_24.columns = ['ref_hash', 'mean_time_to_click']

mean_click_23_25 = clicks_23_25.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_23_25.columns = ['ref_hash', 'mean_time_to_click']

mean_click_24_26 = clicks_24_26.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_click_24_26.columns = ['ref_hash', 'mean_time_to_click']

mean_click_24_26.head()
```

```python
mean_click_18_20['mean_time_to_click'].fillna(0, inplace= True)
mean_click_19_21['mean_time_to_click'].fillna(0, inplace= True)
mean_click_20_22['mean_time_to_click'].fillna(0, inplace= True)
mean_click_21_23['mean_time_to_click'].fillna(0, inplace= True)
mean_click_22_24['mean_time_to_click'].fillna(0, inplace= True)
mean_click_23_25['mean_time_to_click'].fillna(0, inplace= True)
mean_click_24_26['mean_time_to_click'].fillna(0, inplace= True)

mean_click_18_20.head()
```

```python
clicksDT_18_20 = clicksDT_18_20.merge(mean_click_18_20, on='ref_hash', how='left')
clicksDT_19_21 = clicksDT_19_21.merge(mean_click_19_21, on='ref_hash', how='left')
clicksDT_20_22 = clicksDT_20_22.merge(mean_click_20_22, on='ref_hash', how='left')
clicksDT_21_23 = clicksDT_21_23.merge(mean_click_21_23, on='ref_hash', how='left')
clicksDT_22_24 = clicksDT_22_24.merge(mean_click_22_24, on='ref_hash', how='left')
clicksDT_23_25 = clicksDT_23_25.merge(mean_click_23_25, on='ref_hash', how='left')
clicksDT_24_26 = clicksDT_24_26.merge(mean_click_24_26, on='ref_hash', how='left')

clicksDT_19_21.head()
```

# Se agregan la cantidad de sources diferentes.

```python
cant_sources_18_20 = clicks_18_20.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_18_20.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_19_21 = clicks_19_21.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_19_21.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_20_22 = clicks_20_22.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_20_22.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_21_23 = clicks_21_23.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_21_23.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_22_24 = clicks_22_24.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_22_24.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_23_25 = clicks_23_25.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_23_25.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_24_26 = clicks_24_26.groupby('ref_hash')['source_id'].nunique().reset_index()
cant_sources_24_26.columns = ['ref_hash', 'cant_sources_clicks']

cant_sources_20_22.head()
```

```python
cant_sources_18_20['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_19_21['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_20_22['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_21_23['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_22_24['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_23_25['cant_sources_clicks'].fillna(0, inplace= True)
cant_sources_24_26['cant_sources_clicks'].fillna(0, inplace= True)

cant_sources_21_23.head()
```

```python
clicksDT_18_20 = clicksDT_18_20.merge(cant_sources_18_20, on='ref_hash', how='left')
clicksDT_19_21 = clicksDT_19_21.merge(cant_sources_19_21, on='ref_hash', how='left')
clicksDT_20_22 = clicksDT_20_22.merge(cant_sources_20_22, on='ref_hash', how='left')
clicksDT_21_23 = clicksDT_21_23.merge(cant_sources_21_23, on='ref_hash', how='left')
clicksDT_22_24 = clicksDT_22_24.merge(cant_sources_22_24, on='ref_hash', how='left')
clicksDT_23_25 = clicksDT_23_25.merge(cant_sources_23_25, on='ref_hash', how='left')
clicksDT_24_26 = clicksDT_24_26.merge(cant_sources_24_26, on='ref_hash', how='left')

clicksDT_22_24.head()
```

# Se agregan la cantidad de advertiser diferentes.

```python
cant_advs_18_20 = clicks_18_20.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_18_20.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_19_21 = clicks_19_21.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_19_21.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_20_22 = clicks_20_22.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_20_22.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_21_23 = clicks_21_23.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_21_23.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_22_24 = clicks_22_24.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_22_24.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_23_25 = clicks_23_25.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_23_25.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_24_26 = clicks_24_26.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cant_advs_24_26.columns = ['ref_hash', 'cant_advs_clicks']

cant_advs_23_25.head()
```

```python
cant_advs_18_20['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_19_21['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_20_22['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_21_23['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_22_24['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_23_25['cant_advs_clicks'].fillna(0, inplace= True)
cant_advs_24_26['cant_advs_clicks'].fillna(0, inplace= True)

cant_advs_24_26.head()
```

```python
clicksDT_18_20 = clicksDT_18_20.merge(cant_advs_18_20, on='ref_hash', how='left')
clicksDT_19_21 = clicksDT_19_21.merge(cant_advs_19_21, on='ref_hash', how='left')
clicksDT_20_22 = clicksDT_20_22.merge(cant_advs_20_22, on='ref_hash', how='left')
clicksDT_21_23 = clicksDT_21_23.merge(cant_advs_21_23, on='ref_hash', how='left')
clicksDT_22_24 = clicksDT_22_24.merge(cant_advs_22_24, on='ref_hash', how='left')
clicksDT_23_25 = clicksDT_23_25.merge(cant_advs_23_25, on='ref_hash', how='left')
clicksDT_24_26 = clicksDT_24_26.merge(cant_advs_24_26, on='ref_hash', how='left')

clicksDT_18_20.head()
```

# Se agrega el mean location.

```python
mean_loc_18_20 = clicks_18_20.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_18_20.columns = ['ref_hash', 'lon', 'lat']
mean_loc_18_20['mean_location'] = list(zip(mean_loc_18_20['lon'], mean_loc_18_20['lat']))
del mean_loc_18_20['lon']
del mean_loc_18_20['lat']

mean_loc_19_21 = clicks_19_21.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_19_21.columns = ['ref_hash', 'lon', 'lat']
mean_loc_19_21['mean_location'] = list(zip(mean_loc_19_21['lon'], mean_loc_19_21['lat']))
del mean_loc_19_21['lon']
del mean_loc_19_21['lat']

mean_loc_20_22 = clicks_20_22.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_20_22.columns = ['ref_hash', 'lon', 'lat']
mean_loc_20_22['mean_location'] = list(zip(mean_loc_20_22['lon'], mean_loc_20_22['lat']))
del mean_loc_20_22['lon']
del mean_loc_20_22['lat']

mean_loc_21_23 = clicks_21_23.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_21_23.columns = ['ref_hash', 'lon', 'lat']
mean_loc_21_23['mean_location'] = list(zip(mean_loc_21_23['lon'], mean_loc_21_23['lat']))
del mean_loc_21_23['lon']
del mean_loc_21_23['lat']

mean_loc_22_24 = clicks_22_24.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_22_24.columns = ['ref_hash', 'lon', 'lat']
mean_loc_22_24['mean_location'] = list(zip(mean_loc_22_24['lon'], mean_loc_22_24['lat']))
del mean_loc_22_24['lon']
del mean_loc_22_24['lat']

mean_loc_23_25 = clicks_23_25.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_23_25.columns = ['ref_hash', 'lon', 'lat']
mean_loc_23_25['mean_location'] = list(zip(mean_loc_23_25['lon'], mean_loc_23_25['lat']))
del mean_loc_23_25['lon']
del mean_loc_23_25['lat']

mean_loc_24_26 = clicks_24_26.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_loc_24_26.columns = ['ref_hash', 'lon', 'lat']
mean_loc_24_26['mean_location'] = list(zip(mean_loc_24_26['lon'], mean_loc_24_26['lat']))
del mean_loc_24_26['lon']
del mean_loc_24_26['lat']

mean_loc_19_21.head()
```

```python
mean_loc_18_20['mean_location'].fillna(0, inplace= True)
mean_loc_19_21['mean_location'].fillna(0, inplace= True)
mean_loc_20_22['mean_location'].fillna(0, inplace= True)
mean_loc_21_23['mean_location'].fillna(0, inplace= True)
mean_loc_22_24['mean_location'].fillna(0, inplace= True)
mean_loc_23_25['mean_location'].fillna(0, inplace= True)
mean_loc_24_26['mean_location'].fillna(0, inplace= True)

mean_loc_20_22.head()
```

```python
clicksDT_18_20 = clicksDT_18_20.merge(mean_loc_18_20, on='ref_hash', how='left')
clicksDT_19_21 = clicksDT_19_21.merge(mean_loc_19_21, on='ref_hash', how='left')
clicksDT_20_22 = clicksDT_20_22.merge(mean_loc_20_22, on='ref_hash', how='left')
clicksDT_21_23 = clicksDT_21_23.merge(mean_loc_21_23, on='ref_hash', how='left')
clicksDT_22_24 = clicksDT_22_24.merge(mean_loc_22_24, on='ref_hash', how='left')
clicksDT_23_25 = clicksDT_23_25.merge(mean_loc_23_25, on='ref_hash', how='left')
clicksDT_24_26 = clicksDT_24_26.merge(mean_loc_24_26, on='ref_hash', how='left')

clicksDT_21_23.head()
```
