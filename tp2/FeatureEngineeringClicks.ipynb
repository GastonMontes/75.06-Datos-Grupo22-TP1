```python
import pandas as pd
import numpy as np
import datetime
```

```python
#Leo la data proveniente de un csv
path = 'datos/clicks.csv'
clicks = pd.read_csv(path)
```

```python
clicks.head(10)
```

```python
# Damos formato a la fecha.
clicks['timestamp_format'] = pd.to_datetime(clicks['created'])
# Agrego columna fecha
clicks['date_timestamp'] = clicks['timestamp_format'].dt.date
# Agrego columna numero dia.
clicks['day'] = clicks['timestamp_format'].dt.day
```

```python
# ventanas de 3 dias
clicks18_20 = clicks[(clicks['day'] >= 18) & (clicks['day'] < 21)]
clicks19_21 = clicks[(clicks['day'] >= 19) & (clicks['day'] < 22)]
clicks20_22 = clicks[(clicks['day'] >= 20) & (clicks['day'] < 23)]
clicks21_23 = clicks[(clicks['day'] >= 21) & (clicks['day'] < 24)]
clicks22_24 = clicks[(clicks['day'] >= 22) & (clicks['day'] < 25)]
clicks23_25 = clicks[(clicks['day'] >= 23) & (clicks['day'] < 26)]
clicks24_26 = clicks[(clicks['day'] >= 24) & (clicks['day'] < 27)]
```

# Feature Engineering

## Cantidad de Clicks

```python
final = clicks18_20.groupby('ref_hash')['created'].count().reset_index()
final.columns = ['ref_hash', 'cant_clicks']
final2 = clicks21_23.groupby('ref_hash')['created'].count().reset_index()
final2.columns = ['ref_hash', 'cant_clicks']
```

## Clicks con wifi

```python
clicks18_20['wifi_connection'].fillna(0, inplace= True)
clicks18_20['wifi_connection'].astype('int64', inplace= True)
clicks_con_wifi = clicks18_20.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_con_wifi.columns = ['ref_hash', 'cant_wifi_clicks']

final = final.merge(clicks_con_wifi, on='ref_hash', how='left')
final['cant_wifi_clicks'].fillna(0, inplace= True)
```

```python
clicks21_23['wifi_connection'].fillna(0, inplace= True)
clicks21_23['wifi_connection'].astype('int64', inplace= True)
clicks_con_wifi2 = clicks21_23.groupby('ref_hash').agg({'wifi_connection': 'sum'}).reset_index()
clicks_con_wifi2.columns = ['ref_hash', 'cant_wifi_clicks']

final2 = final2.merge(clicks_con_wifi2, on='ref_hash', how='left')
final2['cant_wifi_clicks'].fillna(0, inplace= True)
```

## Promedio de tiempo de clicks

```python
mean_time_to_click = clicks18_20.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_time_to_click.columns = ['ref_hash', 'mean_time_to_click']

final = final.merge(mean_time_to_click, on='ref_hash', how='left')
final['mean_time_to_click'].fillna(0, inplace= True)
```

```python
mean_time_to_click2 = clicks21_23.groupby('ref_hash')['timeToClick'].mean().reset_index()
mean_time_to_click2.columns = ['ref_hash', 'mean_time_to_click']

final2 = final2.merge(mean_time_to_click2, on='ref_hash', how='left')
final2['mean_time_to_click'].fillna(0, inplace= True)
```

## Cantidad de Sources diferentes

```python
cantidad_de_sources = clicks18_20.groupby('ref_hash')['source_id'].nunique().reset_index()
cantidad_de_sources.columns = ['ref_hash', 'cant_sources_clicks']

final = final.merge(cantidad_de_sources, on='ref_hash', how='left')
final['cant_sources_clicks'].fillna(0, inplace= True)
```

```python
cantidad_de_sources2 = clicks21_23.groupby('ref_hash')['source_id'].nunique().reset_index()
cantidad_de_sources2.columns = ['ref_hash', 'cant_sources_clicks']

final2 = final2.merge(cantidad_de_sources2, on='ref_hash', how='left')
final2['cant_sources_clicks'].fillna(0, inplace= True)
```

## Cantidad de Advs diferentes

```python
cantidad_de_advs = clicks18_20.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cantidad_de_advs.columns = ['ref_hash', 'cant_advs_clicks']

final = final.merge(cantidad_de_advs, on='ref_hash', how='left')
final['cant_advs_clicks'].fillna(0, inplace= True)
```

```python
cantidad_de_advs2 = clicks21_23.groupby('ref_hash')['advertiser_id'].nunique().reset_index()
cantidad_de_advs2.columns = ['ref_hash', 'cant_advs_clicks']

final2 = final2.merge(cantidad_de_advs2, on='ref_hash', how='left')
final2['cant_advs_clicks'].fillna(0, inplace= True)
```

## Mean Location

```python
mean_location = clicks18_20.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_location.columns = ['ref_hash', 'lon', 'lat']
mean_location['mean_location'] = list(zip(mean_location['lon'], mean_location['lat']))
del mean_location['lon']
del mean_location['lat']

final = final.merge(mean_location, on='ref_hash', how='left')
final['mean_location'].fillna(0, inplace= True)
```

```python
mean_location2 = clicks21_23.groupby('ref_hash')['longitude', 'latitude'].mean().reset_index()
mean_location2.columns = ['ref_hash', 'lon', 'lat']
mean_location2['mean_location'] = list(zip(mean_location2['lon'], mean_location2['lat']))
del mean_location2['lon']
del mean_location2['lat']

final2 = final2.merge(mean_location, on='ref_hash', how='left')
final2['mean_location'].fillna(0, inplace= True)
```
