```python
# imports de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline

plt.style.use('default')
sns.set(style="whitegrid") # tipo de grid seaborn
```

```python
# Leo el archivo de installs.
installs = pd.read_csv('installs.csv.gzip', compression='gzip')
installs.head(20)

# Leo el archivo de eventos.
events = pd.read_csv('events.csv.gzip', compression='gzip')
events.head(20)
```

```python
installsPerApp = installs[['application_id', 'attributed']]
eventsPerApp = events[['application_id', 'attributed']]

installsPerApp.rename(columns={'attributed': 'attributedInstall'}, inplace=True)
eventsPerApp.rename(columns={'attributed': 'attributedEvent'}, inplace=True)

installsPerApp['installCount'] = 1
eventsPerApp['eventsCount'] = 1


installsPerApp['attributedInstall'] = installsPerApp['attributedInstall'].astype(int)
eventsPerApp['attributedEvent'] = eventsPerApp['attributedEvent'].astype(int)

installsPerApp = installsPerApp.groupby('application_id').sum()
eventsPerApp = eventsPerApp.groupby('application_id').sum()

installsPerApp.reset_index(inplace=True)
eventsPerApp.reset_index(inplace=True)

mergedData = pd.merge(installsPerApp, eventsPerApp, on='application_id', how='outer')
mergedData = mergedData.fillna(0)
mergedData
```

# Top aplicaciones con mas eventos.

```python
eventsPerAppTopCount = mergedData.sort_values(by='eventsCount', ascending=False).head(10)
eventsPerAppTopCount
```

# Top aplicaciones con mas eventos atribuidos a JAMPP.

```python
eventsAttPerAppTopCount = mergedData.sort_values(by='attributedEvent', ascending=False).head(10)
eventsAttPerAppTopCount
```

# Top aplicaciones con mas installs.

```python
installsPerAppTopCount = mergedData.sort_values(by='installCount', ascending=False).head(10)
installsPerAppTopCount
```

# Top aplicaciones con installs atribuídos a JAMPP.

```python
installsAttPerAppTopCount = mergedData.sort_values(by='attributedInstall', ascending=False).head(10)
installsAttPerAppTopCount
```
